```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
import time


backends = ['opencv', 'ssd', 'mtcnn', 'retinaface']

methods = ['HaarCascade', 'SSD', 'MTCNN', 'RetinaFace']

models = ["VGG-Face", "Facenet", "Facenet512", "OpenFace", "DeepFace", "DeepID", "ArcFace", "Dlib"]


_, plots = plt.subplots(1,4, figsize=(15, 15))
for i in range(0, len(backends)):
    tic = time.perf_counter()
    face = DeepFace.detectFace(img_path = "img - copy.jpg", detector_backend = backends[i])
    toc = time.perf_counter()
    plots[i].set_title(f'{methods[i]}\n{toc - tic:0.2f} seconds')
    plots[i].imshow(face)
```
